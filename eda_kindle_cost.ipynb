```python
import pandas as pd
import numpy as np
```

```python
digital_items = pd.read_csv("Digital Items.csv")
```

```python
digital_items.isnull().sum() #WHERE IS THERE MISSING DATA
```

```python
missing = digital_items.isnull().sum()
```

```python
def missing_percent (df, total_missing):  #FUNCTION TO GATHER THE PERCENTAGE OF MISSING DATA TO THE WHOLE
    total = np.product(df.shape)
    print(total)
    total_missing = missing.sum()
    print(total_missing)
    percent_missing = (total_missing / total) * 100
    print(percent_missing)

missing_percent(digital_items, missing)
```
